# Data

you can download the CIFAR-10 dataset in the following link: https://www.cs.toronto.edu/~kriz/cifar.html


```python
#script for opening the files data_batch_x
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict_ = pickle.load(fo, encoding='bytes')
    return dict_

def get_numpy_data_target(file):
    data_dict = unpickle(file)
    X = np.array(list( map(lambda x: x.reshape((32,32,3)),data_dict[b'data']) ) )
    y = data_dict[b'labels']
    return X, y
```

```python
ntest = 5000
Xtrain,y_train = get_numpy_data_target('data/data_batch_1')
Xtest,ytest = get_numpy_data_target('data/data_batch_2')
Xtest = Xtest[:ntest,:,:,:]
ytest = ytest[:ntest]

input_shape = (32,32,3)
n_classes =  10
```

```python
ytrain_to_cat = to_categorical(y_train, num_classes=10)
ytest_to_cat = to_categorical(ytest, num_classes=10)
```

# 1. Alexnet

Convolutional layers combined with max pooling 

```python
Image(filename = "images/alexnet.png", width=600, height=600)
```

```python


def get_AlexNet(input_shape,n_classes):
    model = Sequential()
    model.add(Convolution2D(input_shape = input_shape ,filters=3, kernel_size=(1,1), strides=(1, 1), padding='valid'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Convolution2D(filters=3, kernel_size=(1,1), strides=(1, 1), padding='valid'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    #etc... repeat layers until ...

    model.add(Flatten())
    model.add(Dense(128))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dense(128))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dense(n_classes))
    model.add(BatchNormalization())
    model.add(Activation('softmax'))
    
    model.summary()
    
    return model
```

```python
model = get_AlexNet(input_shape,n_classes)
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss=losses.categorical_crossentropy, optimizer='adam',metrics=['accuracy'])
model.fit(Xtrain,ytrain_to_cat,epochs = 2,validation_data=(Xtest,ytest_to_cat))
```

## VGG Net

Very similar to AlexNet in the use of combining pooling with connvolutional nets. The insight here is to push towards further depths, but narrowing each layer using pooling.

```python
Image(filename = "images/vggnet.png", width=600, height=600)
```

```python
def get_vggNet(input_shape,n_classes):
    model = Sequential()
    model.add(Convolution2D(input_shape = input_shape ,filters=8, kernel_size=(3,3), strides=(1, 1), padding='valid'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Convolution2D(filters=16, kernel_size=(3,3), strides=(1, 1), padding='valid'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Convolution2D(filters=32, kernel_size=(3,3), strides=(1, 1), padding='valid'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))

    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dense(32))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dense(32))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dense(n_classes))
    model.add(BatchNormalization())
    model.add(Activation('softmax'))
    
    model.summary()
    
    return model
```

```python
model = get_vggNet(input_shape,n_classes)
model.compile(loss=losses.categorical_crossentropy, optimizer='adam',metrics=['accuracy'])
model.fit(Xtrain,ytrain_to_cat,epochs = 2,validation_data=(Xtest,ytest_to_cat))
```

## 2. GoogleNet

Introduces the concept of inception modules:

Each layer is a stacking of various operations.

The paper also introduces the idea of Model Regularization via Label Smoothing:



```python
Image(filename = "images/GoogleNet.png", width=900, height=1200)
```

```python
from keras import backend as K
_K = n_classes
_DELTA = 0.1
def cross_entropy_label_smoothing(y_true, y_pred):
    q = (1.-_DELTA)*y_true+_DELTA/_K
    out = losses.categorical_crossentropy(q,y_pred)
    return out
```

```python
#Note this is not the architecture of the paper
#Just an illustration of the implementation in Keras

def get_inception_layer(input_layer):
    conv1 = Conv2D(filters=4, kernel_size=(17, 17), strides=(1,1), activation='relu')(input_layer)
    conv2_0 = Conv2D(filters=4, kernel_size=(9,9), strides=(1,1), activation='relu')(input_layer)
    conv2_1 = Conv2D(filters=4, kernel_size=(9, 9), strides=(1,1), activation='relu')(conv2_0)
    
    conv3_0 = Conv2D(filters=4, kernel_size=(9, 9), strides=(1,1), activation='relu')(input_layer)
    conv3_1 = Conv2D(filters=4, kernel_size=(9, 9), strides=(1,1), activation='relu')(conv3_0)
   
    maxpool4_0 = MaxPooling2D(pool_size=(2, 2))(input_layer)
    conv4_1 = Conv2D(filters=4, kernel_size=(2, 2), strides=(1,1), activation='relu')(maxpool4_0)  
    
    mrg = Concatenate(axis=-1)([conv1,conv2_1,conv3_1,conv4_1])
                              
    return mrg
                              

def get_GoogleNet(input_shape,n_classes):
    inpt = Input(shape=input_shape)
    layer_1 = Conv2D(filters=4, kernel_size=(2,2), strides=(1, 1), padding='valid',activation='relu')(inpt)
    layer_2 = Conv2D(filters=8, kernel_size=(2,2), strides=(1, 1), padding='valid',activation='relu')(layer_1)
    layer_3 = get_inception_layer(layer_2)
    layer_4 = Conv2D(filters=4, kernel_size=(2,2), strides=(1, 1), padding='valid',activation='relu')(layer_3)
    layer_5 = Flatten()(layer_4)
    layer_6 =Dense(32,activation='relu')(layer_5)
    Layer_final = Dense(n_classes,activation='softmax')(layer_6)
    
    model = Model(input=inpt, output=Layer_final)
    model.summary()
    
    return model
```

```python
model = get_GoogleNet(input_shape,n_classes)
model.compile(loss=cross_entropy_label_smoothing, optimizer='adam',metrics=['accuracy'])
model.fit(Xtrain,ytrain_to_cat,epochs = 2,validation_data=(Xtest,ytest_to_cat))
```

# 3. ResNet

source: arXiv:1512.03385v1

ResNet, similarly to GoogleNet, consists stacked layers. However, the output of each stacked layer is also propagated unprocessed. The processed and unprocessed output are then combined by a simple addition.

If $\mathcal{H}(x)$ denotes the underlying model to be fit. Then the stack will be fit on the residual $\mathcal{H}(x)-x$ 

The additional identity mapping was introduced as a technique to counter **degradation**.

As the network depth increases, accuracy gets saturated and then degrades rapidly. Unfortunately such degradation is **not caused by overtting**. 


** Adding more layers to a deep model leads to higher training error. **

Formally the building block is written as:
$$ y = \mathcal{F}\left(x,\{W_i\}\right) + P x $$

where $P$ is an eventual projection for matching the dimension with the output of the stack $\mathcal{F}(x)$. Empirical evidence has shown that the identity/projection is sufficient and replacement with a linear transformation does not add much to the overal performance.

```python
def reduce_dim(x,weight,in_shape=(32,32,3),out_shape=(30,30,2)):
    reshaped_x = Flatten()(x)
    p = K.random_normal(shape=(in_shape[0]*in_shape[1]*in_shape[2],out_shape[0]*out_shape[1]*out_shape[2]) )/np.sqrt(in_shape[0]*in_shape[1]*in_shape[2])
    out_x = K.dot(reshaped_x,p )
    out_x = Reshape(out_shape)(out_x)
    return out_x+weight
```

```python
def resnet_building_block_0(x):
    '''
    Fits a stack F(x) on the input x and returns
    F(x)+x
    '''
    weight_0 = Conv2D(filters=2, kernel_size=(1, 1), strides=(1,1), activation='relu')(x)
    weight_1 = Conv2D(filters=2, kernel_size=(3, 3), strides=(1,1), activation='relu')(weight_0)
    weight_2 = Conv2D(filters=2, kernel_size=(1, 1), strides=(1,1), activation='relu')(weight_1)
    
    out_x = Lambda( lambda x:reduce_dim(x,weight_2),output_shape=(30,30,2))(x)
    return out_x 
def resnet_building_block_1(x):
    '''
    Fits a stack F(x) on the input x and returns
    F(x)+x
    '''
    weight_0 = Conv2D(filters=2, kernel_size=(1, 1), strides=(1,1), activation='relu')(x)
    weight_1 = Conv2D(filters=2, kernel_size=(3, 3), strides=(1,1), activation='relu')(weight_0)
    weight_2 = Conv2D(filters=2, kernel_size=(1, 1), strides=(1,1), activation='relu')(weight_1)
    
    out_x = Lambda( lambda x:reduce_dim(x,weight_2,(30,30,2),(28,28,2)),output_shape=(28,28,2))(x)
    return out_x 
    
def get_ResNet(input_shape,n_classes):
    inpt = Input(shape=input_shape)
    layer_1 = resnet_building_block_0(inpt)
    layer_2 = resnet_building_block_1(layer_1)
    layer_3 = Flatten()(layer_2)
    layer_4 =Dense(2,activation='relu')(layer_3)
    Layer_final = Dense(n_classes,activation='softmax')(layer_4)
    
    model = Model(input=inpt, output=Layer_final)
    model.summary()
    
    return model
```

```python
model = get_ResNet(input_shape,n_classes)
model.compile(loss= losses.categorical_crossentropy, optimizer='adam',metrics=['accuracy'])
model.fit(Xtrain,ytrain_to_cat,epochs = 2,validation_data=(Xtest,ytest_to_cat) )
```
